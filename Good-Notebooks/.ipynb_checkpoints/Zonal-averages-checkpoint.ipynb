```python
for i in np.arange(8,60):
    plt.clf()
    fig = plt.figure(figsize = (10,6))
    (temp_warm - temp_cont).groupby('time.year').mean('time').mean('xt_ocean').isel(year = i).sel(st_ocean = slice(0,2500)).plot(yincrease = False, cmap = cm.cm.balance, vmin = -1.5, vmax = 1.5)
    pot_rho_warm_av = pot_rho_warm.mean('xt_ocean').isel(year = i)
    pot_rho_cont_av = pot_rho_cont.mean('xt_ocean').isel(year = i)
    pot_rho_cont_av_mean = pot_rho_cont.mean('xt_ocean').mean('year')

    clevel = np.arange(1034,1038,0.25)
    cs = plt.contour(pot_rho_warm_av.yt_ocean, pot_rho_warm_av.st_ocean, pot_rho_warm_av, levels=clevel, colors='k', linewidths=0.5)
    cs2 = plt.contour(pot_rho_cont_av.yt_ocean, pot_rho_cont_av.st_ocean, pot_rho_cont_av, levels=clevel, colors='k', linestyle = 'dotted', linewidths=0.25)
    cs3 = plt.contour(pot_rho_cont_av_mean.yt_ocean, pot_rho_cont_av_mean.st_ocean, pot_rho_cont_av_mean, levels=np.array([1035.9901]), colors='magenta', linewidths=1)
    cs4 = plt.contour(pot_rho_warm_av.yt_ocean, pot_rho_warm_av.st_ocean, pot_rho_warm_av, levels=np.array([1035.9901]), colors='orange', linewidths=1)

    plt.ylim(1200,0)
    plt.xlim(-90,0)
    plt.clabel(cs, [1034.0,1035.0,1036.0,1037.0], fmt = '%1.0f')
    plt.savefig('pictures2/isopycnals_'+str(2891+i))
    print(i)
```

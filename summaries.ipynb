## Arguments for cooling anomalies observed in the Southern Ocean

The plot below depicts the zonal average salinity, temperature and age. The thick line is the $\sigma_2^{0c}$ control zero wind stress curl isopycnal, and the thin line is this same isopycnal (1035.9901kgm$^{-3}$) after 50 years of the warming perturbation experiment to illustrate southward isopycnal movement. The age plot in particular shows the upwelling south of $\sigma_2^{0c}$ and downwelling north of it. The salinity plot doesn't have such nice boundaries on it, due to vertical mixing. But there is a clear freshening gradient. (add winter mld??)



This plot decomposes (roughly) the temperature anomaly into a spice (gradient of temperature and salinity along the isopycnal) anomaly and a heave anomaly due to motion of isopycnals. This is for the warming perturbation after 50 years. In the first panel there are two cooling patches. One can be explained by the heave anomaly (third panel, cooling due to enhanced upwelling south of $\sigma_2^{0w}$) and the other by spice anomaly. The spice anomaly comes because isopycnal outcrop moves southward into fresher waters, thus the water at that density must be cooler to be on the isopycnal. It also illustrates the efficiency of isopycnal mixing, and makes sense because it is not visible to be very deep. The spice anomaly is not visible in areas where outcrop salinity doesn't change (near equator) and the southern isopycnals have warming anomalies, poissibly because the isopycnals are moved to more saline water in this region. The heave anomaly has heating on the northern side, consistent with Kial's wind stress curl/upwelling in different places argument and with the analysis of age. There is a bit of overlap, but there is clearly a lot of verticla mixing that is not on isopycnals as seen in the first figure as the zero wind stress curl isopycnal does not draw a clear boundary between regions of up and downwelling.



This plot shows the warming perturbation age anomaly after 50 years. older waters exist south of the $\sigma_2^{0c}$ and younger north of it. The orange line is the new zero wind stress curl isopycnal $\sigma_2^{0w}$. Waters that used to downwell are now in the upwelling region or are upwelling more (to the South) so they have an older anomaly. Note the lack of age anomaly in the lower latitude cooling patch.
